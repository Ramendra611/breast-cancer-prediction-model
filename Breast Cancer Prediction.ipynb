### Defining the models

```python

def models(X_train, y_train):
    
    #logistic regression
    from sklearn.linear_model import LogisticRegression
    log = LogisticRegression(random_state=0,max_iter=5000)
    log.fit(X_train, y_train)
    
    #Decision Tree
    from sklearn.tree  import DecisionTreeClassifier
    tree = DecisionTreeClassifier(criterion = 'entropy',random_state=0)
    tree.fit(X_train,y_train)
    
    #Random forest
    from sklearn.ensemble import RandomForestClassifier
    rand = RandomForestClassifier(random_state=0, n_estimators=10,criterion='entropy')
    rand.fit(X_train,y_train)
    
    #Print the model accuracy
    
    print('Score for Logistic Model: \n', "Train:",log.score(X_train, y_train),"\n","Test:",log.score(X_test, y_test))
    print('\nScore for DecisionTree Model: \n ',"Train:",tree.score(X_train, y_train), "\n","Test:",tree.score(X_test, y_test))
    print('\nScore for RandomForesr Model: \n', "Train:",rand.score(X_train, y_train),"\n","Test:",rand.score(X_test, y_test))
    
```

```python
models(X_train,y_train)
```
